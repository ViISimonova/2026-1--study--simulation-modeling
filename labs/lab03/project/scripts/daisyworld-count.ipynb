# Исследование численности популяций

В данном разделе мы проанализируем, как изменяется количество черных и белых маргариток в течение длительного времени (1000 шагов симуляции).

## 1. Подготовка данных и условий
Для сбора статистики нам нужно определить функции-фильтры, которые будут подсчитывать количество агентов каждого вида на каждом шаге.

````julia
using DrWatson
@quickactivate "project"
using Agents, DataFrames, CairoMakie

# Подключаем логику модели
include(srcdir("daisyworld.jl"))

# Определяем критерии подсчета для сбора данных (adata)
black(a) = a.breed == :black
white(a) = a.breed == :white
adata = [(black, count), (white, count)]

# Инициализируем модель при стандартной светимости солнца
model = daisyworld(; solar_luminosity = 1.0)
````

## 2. Запуск симуляции и сбор статистики
Мы запускаем модель на 1000 шагов. Функция `run!` автоматически соберет данные о численности согласно нашим фильтрам `adata`.

````julia
agent_df, model_df = run!(model, 1000; adata)

# Проверим первые строки полученной таблицы данных
first(agent_df, 5)
````

## 3. Визуализация динамики
Построим график, где по оси X — время (шаги), а по оси Y — количество маргариток. Это позволит увидеть процессы саморегуляции в системе.

````julia
# Создаем полотно для графика
figure = Figure(size = (600, 400))
ax = Axis(figure[1, 1], 
    xlabel = "Время (шаги/ticks)", 
    ylabel = "Количество маргариток",
    title = "Динамика популяций Daisyworld"
)

# Отрисовка линий: черные маргаритки — черным цветом, белые — оранжевым (для контраста)
blackl = lines!(ax, agent_df[!, :time], agent_df[!, :count_black], color = :black)
whitel = lines!(ax, agent_df[!, :time], agent_df[!, :count_white], color = :orange)

# Добавляем легенду для понимания графика
Legend(figure[1, 2], [blackl, whitel], ["Черные", "Белые"], labelsize = 12)

# Отображаем фигуру в отчете
figure
````

## 4. Сохранение результата
Сохраняем финальный график в папку с результатами проекта.

````julia
save(plotsdir("daisy_count.png"), figure)
println("График динамики успешно сохранен в plots/daisy_count.png")
````
